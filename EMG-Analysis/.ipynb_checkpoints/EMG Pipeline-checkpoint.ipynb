## Notebook Helpers

```python
def data_ingestion(num, session):

    data_path = os.path.split(os.getcwd())[0] + '/EMG-Data/P' + f'{num}/P{num}_{session}_data.csv'
    marker_path = os.path.split(os.getcwd())[0] + '/EMG-Data/P' + f'{num}/P{num}_{session}_marker.csv'
    df = pd.read_csv(data_path)
    dfm = pd.read_csv(marker_path)

    return df, dfm
```

```python
def get_full_emg(data_frame):
    
    emg = np.zeros((data_frame.shape[0],7)).T
    
    recording = data_frame.to_numpy()
        
    emg[:,:] = recording[:,1:8].T
    
    print(emg.shape)
    return emg 
```

```python
def filter_emg_wrapper(emg_2D, lowcut, notch_1, notch_2, sample_rate, order, quality):
    
    hp_emg = butter_highpass_filter(emg_2D, lowcut, sample_rate, order)
    
    hp_and_notch_emg = iir_notch_filter(hp_emg, notch_1, quality, sample_rate)
    #hp_and_notches_emg = iir_notch_filter(hp_and_notch_emg, notch_2, quality, sample_rate)
    
    return hp_and_notch_emg
```

## Preprocessing Wrapper Function

```python
def get_2D_preprocessed_array(p_id, session, total_scene_len, n_ch, lowcut, notch_1, notch_2, sample_rate, order, quality):
    df, dfm = data_ingestion(p_id, session)
    emg = get_full_emg(df)
    emg_co = correct_mean_emg(emg, n_ch)
    emg_filtered = filter_emg_wrapper(emg_co, lowcut, notch_1, notch_2, sample_rate, order, quality)
    emg_ecg_rem =  wavelet_denoise(emg_filtered, n_ch)
    return emg_ecg_rem, df, dfm
```

## Epoching (Data Segmentation)

```python
def get_epochs(emg_2D, df, dfm, fs):
    # Establishing what vocabulary words are present
    sceneDict = dict()
    
    #reformats OpenBCI timestamps so we can compare with marker timestamps
    for i in range(0,int(df.shape[0])):
        df['Time'][i] = int(df['Time'][i]*1000)

    # Adding row keys of relevant frequencies from dataframe
    for i, scene_code in enumerate(dfm[' SceneNum']): #theres an important space here

        if not np.isnan(scene_code):

            #searches for row in data corresponding to marker
            row = df[df['Time'] < dfm['TimeStamps'][i]].index[-1]

            #gets last instance of that scene if the participant replayed, otherwise gets only instance of scene
            sceneDict.update({scene_code: row})
            
    sceneArray = np.array(list(sceneDict.keys()))
    
    # Scene COUNT (for 4D array later)
    num_scenes = len(sceneArray)

    n_samples = int(df.shape[0])  #full number of rows in csv

    #instantiates emg data in 3 dimensional np array
    emg_3D = np.zeros((num_scenes, n_ch, total_scene_len))

    #instantiates time data in 3 dimensional np array
    timestamps = np.zeros(shape=(num_scenes, n_ch, total_scene_len))

    # Assigning the correct values to the matrix/object
    start_idx_list = []

    #grabs start and endpoints for each Hockey Scene
    for i, scene in enumerate(sceneArray):
        
            start_idx = int(sceneDict[scene]) + 10*int(fs) #offsets to disregard initial 10s wait

            #print(start_idx)
            start_idx_list.append(start_idx)
            end_idx = start_idx + total_scene_len

            #np array for timestamps
            timestamps[i, :, :] = np.array(df.iloc[start_idx:end_idx, 0:1]).transpose((1, 0))

            #shaves off timestamps and markers and does a transpose, we transpose it back and cast as a np array
            emg_3D[i, :, :] = emg_2D[:,start_idx:end_idx]

    return emg_3D, timestamps, sceneArray
```

```python
# FFt of Ch1
fftData = np.fft.fft(ch1)
freq = np.fft.fftfreq(len(ch1))*250 #250h

# Remove unnecessary negative reflection
fftData = fftData[1:int(len(fftData)/2)]
freq = freq[1:int(len(freq)/2)]

# Recall FFT is a complex function
fftData = np.sqrt(fftData.real**2 + fftData.imag**2)

# Plot for sanity check
plt.plot(freq, fftData)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
plt.clf()
```

```python
# We want time information as well as frequency information so FFT's wont help us here
# We also need to filter the EMG signal 


# let x be our optimal signal, then let y = x + sigma*E be our real signal where E is the noise and sigma 
# is the noise amplitude that scales the noise

# using the scikit image library we can use a discrete wavelet transformation to denoise the signal (1d image) by
# decomposing the signal into its coefficients and then applying a threshold to those coeficients before the 
# IWT is applied to reconstruct it

ch1 = np.array(ch1)


# soft universal thresholding method applied
# wavelet decomposes signal by 6 levels (6 detail and 6 approximate coefficients)
# once threshold is applied, signal gets reconstructed, this all happens in one function call
ch1_filtered_db = denoise_wavelet(ch1, method='VisuShrink', mode='soft',wavelet_levels=6,wavelet='db7',rescale_sigma=True)


# Plot filtered Ch1 against time 
plt.plot(time, ch1_filtered_db)
plt.xlabel("Unix Time")
plt.ylabel("Î¼V")
plt.show()
plt.clf()
```
